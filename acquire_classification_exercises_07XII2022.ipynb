**Nº 3 / Make a function named get_telco_data that returns the data from the telco_churn database in SQL.**  
In your SQL, be sure to join contract_types, internet_service_types, payment_types tables with the customers table, so that the resulting dataframe contains all the contract, payment, and internet service options.

```python
def get_telco_data(teld):
    
    if os.path.isfile('file.csv'):
        
        return pd.read_csv('file.csv')
    
    else:
        
        url = teld('telco_churn')
    
        query = '''
        SELECT * 
        FROM customers
        JOIN contract_types 
            USING (contract_type_id)
        JOIN internet_service_types
            USING (internet_service_type_id)
        JOIN payment_types
            USING (payment_type_id)
                '''
        telco_df = pd.read_sql(query, url)
        
        telco_df.to_csv('telco_churn.csv')
        
    return telco_df
    
```
